```python
#geometry=Rectangle(dl.Point(0.0, 0.0), dl.Point(1, 1))
#mesh = (generate_mesh(geometry,40))

mesh=mshr.generate_mesh(mshr.Circle(Point(0.,0.),1.),40)

#mesh = dl.refine(UnitSquareMesh(30,30))

V = FunctionSpace(mesh, 'P', 2)

# Define boundary condition
tol = 1E-14
u_D = Expression('0', degree=2)

def boundary(x, on_boundary):
    return on_boundary


#def boundary_L(x):
  #  return  near(x[0]**2+x[1]**2, 1.0, tol) 

bc = DirichletBC(V, u_D, boundary)


bcs = [ bc]
```

```python
u = TrialFunction(V)
v = TestFunction(V)
f = Expression('exp(-100*(pow(x[0], 2) + pow(x[1] , 2)))',degree=2)

f=interpolate(f,V)


A1=Constant(((10.0,0.0),(0.0,10.0)))

A2=Constant(((1,-5),(-5,100)))

a = dot(A1*grad(u), grad(v))*dx
L = f*v*dx

# Compute solution

u = Function(V)


solve(a == L, u, bcs)



#A,b = assemble_system(a,L,bcs)

#U = u.vector()
#solve(A,U,b)

# Plot solution and mesh
c=plot(u,mode='color',title="Corresponding to A1 tensor")
plt.colorbar(c)
plt.show()
```

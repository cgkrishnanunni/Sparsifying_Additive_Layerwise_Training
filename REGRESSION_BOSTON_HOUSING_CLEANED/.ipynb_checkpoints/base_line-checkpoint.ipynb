```python
# BASELINE PERFORMANCE, EARLY STOPPING ENFORCED



import sklearn.model_selection as sk
from sklearn.datasets import load_boston
from sklearn.datasets import make_regression
from sklearn.datasets import make_friedman1
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing 
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot 
```

```python
data_train=np.loadtxt("data_train.data")
labels_train=np.loadtxt("labels_train.data")
data_test=np.loadtxt("data_test.data")
labels_test=np.loadtxt("labels_test.data")
```

```python
x_train=data_train
y_train=labels_train
x_test=data_test
y_test=labels_test
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

val=[]
for i in range(0,20):
    
    callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=200,restore_best_weights=True)
    model = keras.Sequential()
    model.add(layers.Dense(100, activation='linear', input_shape=(13,)))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(100, activation='elu'))
    model.add(layers.Dense(1, activation='linear'))
    model.summary()

    opt = keras.optimizers.Adam(learning_rate=0.001)
    model.compile(loss='mean_squared_error', optimizer=opt, metrics=['mean_squared_error'])
    history=model.fit(x_train,y_train,batch_size=100,epochs=10000,validation_data=[x_test, y_test],callbacks=[callback],verbose=1)
    val=np.append(val,model.evaluate(x_test, y_test)[0])
```
